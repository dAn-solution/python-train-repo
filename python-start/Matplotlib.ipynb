## plt.figure()

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(x, x**2)
ax2.plot(x, x**3)
```

## plt.subplots()

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x**2)
axes[1].plot(x, x**3)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0, 0].plot(x, x**2)
axes[0, 1].plot(x, x**4)
axes[1, 0].plot(x, x**3)
axes[1, 1].plot(x, x**5)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, x**2, label='x**2')
axes[1].plot(x, x**3, label='x**3')
axes[0].set_xlabel('xlabel1')
axes[1].set_xlabel('xlabel2')
axes[0].set_ylabel('ylabel1')
axes[1].set_ylabel('ylabel2')
axes[0].set_title('title1')
axes[1].set_title('title2')
axes[0].set_xticks([-3, -2, -1, 5])
axes[0].legend()
axes[1].legend()
axes[1].axis('off')
fig.tight_layout()
```

## plt.scatter()

```python
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
plt.scatter(df['revenue'], df['budget'], alpha=0.2)
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
df['revenue'] = df['revenue'].apply(lambda x: np.nan if x==0 else x)
df = df.dropna(subset=['revenue'])
df['budget'] = df['budget'].apply(lambda x: np.nan if x==0 else x)
df = df.dropna(subset=['budget'])
lang_group_df = df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
c = np.arange(len(lang_group_df))
plt.scatter('revenue', 'budget', s='vote_count', c = c, data=lang_group_df)
for idx, row in lang_group_df.iterrows():
    plt.annotate(idx, (row['revenue'], row['budget']))
```

## plt.hist()

```python
df = pd.read_csv('tmdb_5000_movies.csv')
plt.hist(df['vote_average'], bins=50)
plt.show()
```

```python
## profit(revenue - budget)
df = pd.read_csv('tmdb_5000_movies.csv')
df['revenue'] = df['revenue'].apply(lambda x: 0 if np.isnan(x) else x)
df['budget'] = df['budget'].apply(lambda x: 0 if np.isnan(x) else x)
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
df['profit'] = df['revenue'] - df['budget']
plt.hist(df['profit'])
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
fig, axes = plt.subplots()
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
df['profit'] = df.apply(lambda x: x['revenue'] - x['budget'], axis=1)
axes.hist(df['profit'], bins=50)
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
fig, axes = plt.subplots()
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
df['profit'] = df.apply(lambda x: x['revenue'] - x['budget'], axis=1)
logbins = np.logspace(5,np.log10(df['profit'].max()),100)
axes.hist(df['profit'], bins=logbins)
axes.set_xscale('log')
plt.show()
```

```python
df = pd.read_csv('tmdb_5000_movies.csv')
plt.scatter('runtime', 'budget', data=df, alpha=0.2)
plt.xlabel('runtime')
plt.ylabel('budget')
plt.title('runtime_budget')
plt.savefig('runtime_budget_scatter.png')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 3))
x = np.linspace(-3, 3, 100)
axes[0].plot(x, np.cos(x))
axes[1].scatter(df['revenue'], df['budget'])
axes[0].set_xlabel('x')
axes[0].set_ylabel('cos(x)')
axes[0].set_title('x VS cos(x)')
axes[1].set_xlabel('revenue')
axes[1].set_ylabel('budget')
axes[1].set_title('revenue VS budget')
fig.tight_layout()
fig.savefig('x_cos_revenue_budget_2*2')
```

## PDF

```python
from matplotlib.backends.backend_pdf import PdfPages

pdf = PdfPages('pdf_files.pdf')

fig, axes = plt.subplots()
axes.plot(x, x**2)

pdf.savefig(fig)
pdf.close()
```
