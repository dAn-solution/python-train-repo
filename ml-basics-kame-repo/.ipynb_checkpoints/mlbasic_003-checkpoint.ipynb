## かめさん講座　機械学習入門
#### 6.線型回帰の係数の解釈（p値）

```python
# import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from scipy import stats
from sklearn.linear_model import LinearRegression
```

```python
# dataset
df = sns.load_dataset('diamonds')
df = df[(df[['x', 'y', 'z']] != 0).all(axis=1)]
X = df['depth'].values
y = df['price'].values
```

```python
# m✕nの形にする
X = X.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# 学習する
model = LinearRegression()

model.fit(X, y)
```

```python
# 学習済みモデルのパラメータを確認する
print(model.coef_, model.intercept_)
```

```python
# 切片の係数１を含めた行列
X_ = np.append(np.ones((len(X), 1)), X, axis=1)
# 線形回帰のパラメータ
theta = np.append(model.intercept_, model.coef_)
# 各データの予測値
y_preds = model.predict(X)
```

```python
# RSS
RSS = np.sum((y - y_preds)**2)
# RSE
RSE = np.sqrt(RSS / (len(X) - len(X_[0])))
# 標準誤差
SE_sq = RSE**2 * np.linalg.inv(np.dot(X_.T,X_)).diagonal()
```

```python
# t値を計算
t = theta / np.sqrt(SE_sq)
print(t)
```

```python
# p値の計算
p = [2*(1-stats.t.cdf(np.abs(t_val),(len(X_)-len(X_[0])))) for t_val in t]
print(p)
```

```python
# depth と price の相関
sns.scatterplot(x=df['depth'], y=df['price'], alpha=0.1)
```

#### statsmodelsを使ってp値を確認する

```python
import statsmodels.api as sma
X2 = sma.add_constant(X)
est = sma.OLS(y, X2)
est_trained = est.fit()
print(est_trained.summary())
```

```python
# 複数の特徴量を使ってのp値の確認
# import statsmodels.api as sma
X = df[['carat', 'depth', 'table', 'x', 'y', 'z']].values
X2 = sma.add_constant(X)
est = sma.OLS(y, X2)
est_trained = est.fit()
print(est_trained.summary())
```
