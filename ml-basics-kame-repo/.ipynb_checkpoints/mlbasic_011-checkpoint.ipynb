```python
# import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
# dataset
df = sns.load_dataset('iris')
df.head()
```

```python
df.info()
```

```python
df['species'].unique()
```

```python
df.describe()
```

```python
plt.plot(df['sepal_length'], label='s_length')
plt.plot(df['sepal_width'], label='s_width')
plt.plot(df['petal_length'], label='p_length')
plt.plot(df['petal_width'], label='p_width')
plt.legend()
plt.show()
```

```python
sns.pairplot(df, hue='species')
```

```python
# 学習データとテストデータ作成
X = df.loc[:, df.columns!='species']
y = df['species']
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
print(len(X_train), len(X_test))
```

```python
# 学習
model = LogisticRegression(penalty='none')
model.fit(X_train, y_train)
```

```python
X_test.head()
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
```

```python
# Accuracy
accuracy_score(y_pred, y_test)
```

```python
# Precision
# None
print(precision_score(y_pred, y_test, average=None))
# macro
print(precision_score(y_pred, y_test, average='macro'))
# micro
print(precision_score(y_pred, y_test, average='micro'))
```

```python
# Recall
# None
print(recall_score(y_pred, y_test, average=None))
# macro
print(recall_score(y_pred, y_test, average='macro'))
# micro
print(recall_score(y_pred, y_test, average='micro'))
```

```python
# Specificity
res = []
for c in model.classes_:
    res.append(recall_score(np.array(y_pred)!=c, np.array(y_test)!=c))
res
```

```python
print(np.mean(res))
```

```python
# F-score
print(f1_score(y_pred, y_test, average='macro'))
```
