#### 24.次元削減　ＰＣＡ（主成分分析）

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# 標準化
X = df.loc[:, df.columns!='species']
y = df['species'].values
X = StandardScaler().fit_transform(X)
```

```python
# PCA
pca = PCA(n_components=2)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
y = y.reshape(-1, 1)
df_pc = pd.DataFrame(np.concatenate([X_pca, y], axis=1), \
                     columns=['first component', 'second component', 'species']). \
                    astype({'first component':float, 'second component':float})
```

```python
sns.scatterplot(x='first component', y='second component', hue='species', data=df_pc)
```

```python
# 標準化
X = df.loc[:, df.columns!='species']
y = df['species'].values
X = StandardScaler().fit_transform(X)
```

```python
# PCA
pca = PCA(n_components=3)
pca.fit(X)
X_pca = pca.transform(X)
```

```python
y = y.reshape(-1, 1)
df_pc = pd.DataFrame(np.concatenate([X_pca, y], axis=1), \
                     columns=['first component', 'second component', 'third component', 'species']). \
                    astype({'first component':float, 'second component':float, 'third component':float})
```

```python
sns.scatterplot(x='first component', y='second component', hue='species', data=df_pc)
```

```python
sns.scatterplot(x='first component', y='third component', hue='species', data=df_pc)
```

```python
sns.scatterplot(x='third component', y='second component', hue='species', data=df_pc)
```

```python
# 寄与率
pca = PCA(n_components=4)
pca.fit(X)
pca.explained_variance_ratio_
```

```python
# 累積寄与率
np.cumsum(pca.explained_variance_ratio_)
```

```python
n_components = 4
plt.plot(range(1, n_components+1), np.cumsum(pca.explained_variance_ratio_))
plt.xticks(range(1, n_components+1))
plt.xlabel('components')
plt.ylabel('cumulative explained variance')
```
