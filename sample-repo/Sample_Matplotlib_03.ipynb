```python
# import
import numpy as np
import matplotlib.pyplot as plt
```

#### plot図

```python
x = np.arange(-10, 11)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(x, y, 'o')
plt.show()
```

```python
plt.plot(x, y, 'o-')
plt.show()
```

```python
plt.plot(x, y, 'ro-')
plt.show()
```

#### タイトル、ラベル、軸範囲

```python
plt.plot(x, y)
plt.title('Graph title', size=20, color='green')
plt.xlabel('X label', size=15)
plt.ylabel('Y label', size=15)

plt.xlim(-15, 15)
plt.ylim(-50, 150)
plt.show()
```

#### 複数線の表示

```python
# 複数の曲線を表示
x = np.arange(-10, 11)
y1 = x**2
y2 = (x-1)**2
y3 = (x-2)**2

plt.plot(x, y1, 'r', label='x**2')
plt.plot(x, y2, 'b', label='(x-1)**2')
plt.plot(x, y3, 'y', label='(x-2)**2')
plt.legend()
plt.show()
```

#### 複数のplotを並べて表示

```python
x = np.arange(-10, 11)
y = x**2
```

```python
plt.subplot(131)
plt.plot(x, y, 'r')
plt.title('Graph 1-1')

plt.subplot(132)
plt.plot(x, y, 'b')
plt.title('Graph 1-2')

plt.subplot(133)
plt.plot(x, y, 'y')
plt.title('Graph 1-3')

plt.show()
```

```python
plt.subplot(221)
plt.plot(x, y, 'r')
plt.title('Graph 1-1')

plt.subplot(222)
plt.plot(x, y, 'b')
plt.title('Graph 1-2')

plt.subplot(223)
plt.plot(x, y, 'y')
plt.title('Graph 2-1')

plt.subplot(224)
plt.plot(x, y, 'g')
plt.title('Graph 2-1')

plt.tight_layout() # グラフが重ならないように自動調整

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.plot(x, y, 'r')
plt.title('Graph 1-1')

plt.subplot(222)
plt.plot(x, y, 'b')
plt.title('Graph 1-2')

plt.subplot(223)
plt.plot(x, y, 'y')
plt.title('Graph 2-1')

plt.subplot(224)
plt.plot(x, y, 'g')
plt.title('Graph 2-1')

plt.tight_layout() # グラフが重ならないように自動調整

plt.show()
```
