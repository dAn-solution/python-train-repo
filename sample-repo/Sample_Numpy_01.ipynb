```python
# import
import numpy as np
```

```python
# 初期化
a = np.zeros((2,3))
a
```

```python
# list -> ndarray
a = [10, 20, 30, 40, 40, 50, 20, 50, 20, 30]
b = np.array(a)
a
```

```python
# mean,max,min,median
print(f"平均：{np.mean(a)}")
print(f"最大値：{np.max(a)}")
print(f"最小値：{np.min(a)}")
print(f"中央値：{np.median(a)}")
print(f"合計値：{np.sum(a)}")
```

```python
a = np.arange(10,51)
print(a)
print(np.sort(a))
print(a[::-1])
print(np.arange(50, 9, -1))
```

```python
a = np.arange(0, 12).reshape(3, 4)
print(np.count_nonzero(a < 4))
print(np.count_nonzero(a < 4, axis=0))
print(np.count_nonzero(a < 4, axis=1))
```

```python
print(np.any(a < 4))
print(np.any(a > 100))
print(np.any(a < 4, axis=0))
print(np.any(a > 100, axis=1))
```

```python
print(np.all(a < 4))
print(np.all(a < 100))
print(np.all(a < 4, axis=1))
print(np.all(a < 100, axis=0))
```

```python
print(np.where(a < 5, 44, 99))
print(np.where(a < 5, 44, a))
```

```python
print(a)
a_del = np.delete(a, 1, axis=0)
print(a_del)
a_del = np.delete(a_del, 2, axis=1)
print(a_del) 
a_del = np.delete(a, 2, axis=None)
print(a_del)
```

```python
print(a)
print(np.delete(a, [1, 3], axis=1))
print(np.delete(a, [0, 2], axis=0).reshape(4,))
```
