### 複数グラフ

```python
# import
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x1 = np.arange(0, 10)
y1 = np.random.randint(0, 10, 10)
p11 = np.polyfit(x1, y1, deg=1)
f11 = np.poly1d(p11)
p12 = np.polyfit(x1, y1, deg=3)
f12 = np.poly1d(p12)

x2 = np.arange(0, 10)
y21 = np.random.randint(300, 500, 10)
y22 = np.random.randint(120, 520, 10)

x3 = np.linspace(0, 10, 1000)
y31 = np.sin(x3)
y32 = np.cos(x3)

x4 = np.linspace(-5, 5, 501)
y41, y42, y43 = [], [], []
for i in range(len(x4)):
    y41.append(norm.pdf(x4[i], loc=0, scale=1))
    y42.append(norm.pdf(x4[i], loc=1, scale=1))
    y43.append(norm.pdf(x4[i], loc=-2, scale=2))

x5 = np.arange(0, 10)
y51 = np.random.randint(1, 10, 10)
y52 = np.random.randint(0, 10, 10)

y61 = np.random.normal(loc=15, scale=3, size=1000)
y62 = np.random.normal(loc=21, scale=2, size=1000)
y63 = np.random.normal(loc=12, scale=2, size=1000)

labels7 = ['E', 'D', 'C', 'B', 'A']
y7 = [10, 20, 30, 40, 50]
ex = [0, 0, 0, 0, 0.1]
```

```python
fig = plt.figure(figsize=(14, 15))

ax1 = fig.add_subplot(331, xlabel='x-label', ylabel='y-label', title='scatter')
ax1.scatter(x1, y1, color='b', label='x1-y1')
ax1.plot(x1, f11(x1), color='r', label='plot')
ax1.plot(x1, f12(x1), color='y', label='p;pt3')
ax1.legend()

ax2 = fig.add_subplot(332, title='bar')
ax2.bar(x2, y21, color='y', label='x2-y21')
ax2.plot(x2, y22, color='r', label='x2-y22')
ax2.legend()

ax3 = fig.add_subplot(333, title='sin-cos')
ax3.plot(x3, y31, color='r', label='sin')
ax3.plot(x3, y32, color='b', label='cos')
ax3.legend()

ax4 = fig.add_subplot(334, title='norm')
ax4.plot(x4, y41, color='r', label='norm0')
ax4.plot(x4, y42, color='b', label='norm1')
ax4.plot(x4, y43, color='y', label='norm-2')
ax4.legend()

ax5 = fig.add_subplot(335, title='hist')
ax5.hist(y51, color='r', label='hist')
ax5.plot(x5, y52, color='b', label='plot')
ax5.legend()

ax6 = fig.add_subplot(336, title='boxplot')
ax6.boxplot([y61, y62, y63], labels=['left', 'center', 'right'])
ax6.set_xlabel('xlabel')
ax6.set_ylabel('ylabel')
ax6.set_ylim(0, 30)

ax7 = fig.add_subplot(337, title='pie')
ax7.pie(y7, explode=ex, labels=labels7, autopct='%1.1f%%', startangle=90)

plt.subplots_adjust(wspace=0.2, hspace=0.4)

plt.show()
```

#### 標準正規分布、t分布

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
t1 = stats.t(1).pdf(x)
plt.plot(x, y, c="r", label="norm")
plt.plot(x, t1, c="b", label="dof=1")
plt.legend()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
for dof in np.arange(1, 12, 5):
    t = stats.t(dof).pdf(x)
    plt.plot(x, t, label=(f'dof={dof}'))
plt.plot(x, y, label='norm')
plt.legend()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
fig = plt.figure(figsize=(20, 18))
ax = fig.add_subplot(331, xlabel='x-label', ylabel='y-label', title='norm t')
ax.plot(x, y, c = "r", label='norm')
coler = ["g", "y", "b"]
coler_cnt = 0

for dof in np.arange(1, 12, 5):
    t = stats.t(dof).pdf(x)
    ax.plot(x, t, c=coler[coler_cnt], label=(f'dof={dof}'))
    coler_cnt += 1
ax.legend()
plt.show()
```

```python
def data_search(df, dai_bunrui):
    df = df[(df['対象月'] >= 202006) & (df['対象月'] <= 202112)]
    df = df[df['大分類'] == dai_bunrui]
    df = df.groupby('対象月').sum()
    return df
```

```python
fig = plt.figure(figsize=(15, 12))
ax1 = [] ; ax2 = []
for idx, dai in enumerate(dai_list):
    df_graph = data_search(df, dai)
    left = np.arange(0, df_graph.shape[0])

    height = df_graph['期首在庫金額']/1000
    height1 = df_graph['期末在庫金額']/1000
    y1 = df_graph['販売金額']/1000
    y2 = df_graph['仕入金額']/1000
    y4 = df_graph['粗利額']/1000
    labels = df_graph.index
    width = 0.3

    # 第一軸(ax1)と第二軸(ax2)を作って関連付ける
    ax1.append(fig.add_subplot(4,2,idx+1))
    ax2.append(ax1[idx].twinx())

    #第一軸を折れ線グラフ、第二軸を棒グラフに
    ax1[idx].bar(left - width / 2, height, color='g', label='期首在庫金額', tick_label=labels, width=width)
    ax1[idx].bar(left + width / 2, height1, color='pink', label='期末在庫金額', width=width)

    ax2[idx].plot(left, y1, linewidth=2, color="red", linestyle="solid", marker="o", markersize=8, label='販売金額')
    ax2[idx].plot(left, y2, linewidth=2, color="blue", linestyle="solid", marker="o", markersize=8, label='仕入金額')
    # ax2.plot(left, y3, linewidth=2, color="blue", linestyle="solid", marker="o", markersize=8, label='原価')
    ax2[idx].plot(left, y4, linewidth=2, color="black", linestyle="solid", marker="o", markersize=8, label='粗利額')

    plt.title('仕入商品粗利情報 ' + str(idx), fontsize=13)
    fig.text(0.9, 0.9, '単位：千円', ha='right', va='top')

    if idx == 5:
        ax1[idx].legend(loc='upper center')
        ax2[idx].legend(loc='best')
    ax1[idx].grid()
    ax1[idx].set_xticks(left)
    ax1[idx].set_xticklabels(["", "", ""])
    ax1[idx].set_ylim(0, df_graph['期首在庫金額'].max()/1000)
    ax2[idx].set_ylim(min([df_graph['販売金額'].min()/1000, df_graph['仕入金額'].min()/1000, df_graph['粗利額'].min()/1000]), 
                      max([df_graph['販売金額'].max()/1000, df_graph['仕入金額'].max()/1000, df_graph['粗利額'].max()/1000]))
    # ax1[idx].set_yticklabels([0, (df_graph['期首在庫金額'].max()/2)/1000, df_graph['期首在庫金額'].max()/1000])

    ax1[idx].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
    ax2[idx].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

    if idx >= 6:
        ax1[idx].set_xticklabels(labels)
        fig.autofmt_xdate(rotation=45)
plt.tight_layout() 
plt.show()
```
