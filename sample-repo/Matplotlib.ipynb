```python
class ADD():
    def __init__(self, value):
        self.value = value
    def __add__(self, other):
        if other.value == 20:
            other.value = 3
        return self.value + other.value + 1
```

```python
class SUB():
    def __init__(self, value):
        self.value = value
    def __sub__(self, other):
        if other.value == 5:
            other.value = 25
        return self.value - other.value - 1
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
t1 = stats.t(1).pdf(x)
plt.plot(x, y, c="r", label="norm")
plt.plot(x, t1, c="b", label="dof=1")
plt.legend()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
for dof in np.arange(1, 12, 5):
    t = stats.t(dof).pdf(x)
    plt.plot(x, t, label=(f'dof={dof}'))
plt.plot(x, y, label='norm')
plt.legend()
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x)
fig = plt.figure(figsize=(20, 18))
ax = fig.add_subplot(331, xlabel='x-label', ylabel='y-label', title='norm t')
ax.plot(x, y, c = "r", label='norm')
coler = ["g", "y", "b"]
coler_cnt = 0

for dof in np.arange(1, 12, 5):
    t = stats.t(dof).pdf(x)
    ax.plot(x, t, c=coler[coler_cnt], label=(f'dof={dof}'))
    coler_cnt += 1
ax.legend()
plt.show()
```

```python
def data_search(df, dai_bunrui):
    df = df[(df['対象月'] >= 202006) & (df['対象月'] <= 202112)]
    df = df[df['大分類'] == dai_bunrui]
    df = df.groupby('対象月').sum()
    return df
```

```python
dai_bunrui_list = ['人工歯', '化工品', '模型', '器材', 'ネイル', 'アイキャスト化工品', 'アイキャスト器材', 'サンメディカル']
dai_list = ['111_人工歯', '112_化工品', '113_模型', '114_器材', '115_ネイル','1162_アイキャスト化工品', '1164_アイキャスト器材', '117_サンメディカル']
 
display_idx = [421,422,423,424,425,426,427,428]
```

```python
dai_list
```

```python
fig = plt.figure(figsize=(15, 12))
ax1 = [] ; ax2 = []
for idx, dai in enumerate(dai_list):
    df_graph = data_search(df, dai)
    left = np.arange(0, df_graph.shape[0])

    height = df_graph['期首在庫金額']/1000
    height1 = df_graph['期末在庫金額']/1000
    y1 = df_graph['販売金額']/1000
    y2 = df_graph['仕入金額']/1000
    y4 = df_graph['粗利額']/1000
    labels = df_graph.index
    width = 0.3

    # 第一軸(ax1)と第二軸(ax2)を作って関連付ける
    ax1.append(fig.add_subplot(4,2,idx+1))
    ax2.append(ax1[idx].twinx())

    #第一軸を折れ線グラフ、第二軸を棒グラフに
    ax1[idx].bar(left - width / 2, height, color='g', label='期首在庫金額', tick_label=labels, width=width)
    ax1[idx].bar(left + width / 2, height1, color='pink', label='期末在庫金額', width=width)

    ax2[idx].plot(left, y1, linewidth=2, color="red", linestyle="solid", marker="o", markersize=8, label='販売金額')
    ax2[idx].plot(left, y2, linewidth=2, color="blue", linestyle="solid", marker="o", markersize=8, label='仕入金額')
    # ax2.plot(left, y3, linewidth=2, color="blue", linestyle="solid", marker="o", markersize=8, label='原価')
    ax2[idx].plot(left, y4, linewidth=2, color="black", linestyle="solid", marker="o", markersize=8, label='粗利額')

    plt.title('仕入商品粗利情報 ' + dai, fontsize=13)
    fig.text(0.9, 0.9, '単位：千円', ha='right', va='top')

    if idx == 5:
        ax1[idx].legend(loc='upper center')
        ax2[idx].legend(loc='best')
    ax1[idx].grid()
    ax1[idx].set_xticks(left)
    ax1[idx].set_xticklabels(["", "", ""])
    ax1[idx].set_ylim(0, df_graph['期首在庫金額'].max()/1000)
    ax2[idx].set_ylim(min([df_graph['販売金額'].min()/1000, df_graph['仕入金額'].min()/1000, df_graph['粗利額'].min()/1000]), 
                      max([df_graph['販売金額'].max()/1000, df_graph['仕入金額'].max()/1000, df_graph['粗利額'].max()/1000]))
    # ax1[idx].set_yticklabels([0, (df_graph['期首在庫金額'].max()/2)/1000, df_graph['期首在庫金額'].max()/1000])

    ax1[idx].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
    ax2[idx].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

    if idx >= 6:
        ax1[idx].set_xticklabels(labels)
        fig.autofmt_xdate(rotation=45)
plt.tight_layout() 
plt.show()
```
