```python
# word_count
def word_count():
    """ turtle.txtにある単語毎の出現数をカウントする関数
        特殊文字は除外する
    """
    # turtle.txt read
    f = open("beginner/word-count-challenge/turtle.txt", "r")
    data = f.read()
    f.close()
    # word count
    exception_word = '(),!?"'
    words_dict = {}
    for word in data.lower().split():
        if word.isalpha():
            temp_word = word
        else:
            temp_word = ""
            for j in list(word):
                if not(j in exception_word):
                    temp_word += j
        words_dict[temp_word] = words_dict.get(temp_word, 0) + 1
    words_dict = sorted(words_dict.items(), key=lambda x:x[1], reverse=True)
    return words_dict


turtle_dict = word_count()

# word , count 
print("word , count")
for word in turtle_dict:
    print("{} , {}".format(word[0], word[1]))
```

```python
# word_count(collections, inflector)
def word_count():
    """ turtle.txtにある単語毎の出現数をカウントする関数
        特殊文字は除外する
        collectionsモジュールでカウント
        inflectorで複数形を単数形に
    """
    from collections import Counter
    from inflector import Inflector
    # turtle.txt read
    f = open("beginner/word-count-challenge/turtle.txt", "r")
    data = f.read()
    f.close()
    # word count
    exception_word = '(),!?"'
    words_list = []
    for word in data.lower().split():
        if word.isalpha():
            temp_word = word
        else:
            temp_word = ""
            for j in list(word):
                if not(j in exception_word):
                    temp_word += j
        words_list.append(Inflector().singularize(temp_word))
        
    words_dict = dict(Counter(words_list))
    words_tuple = sorted(words_dict.items(), key=lambda x:x[1], reverse=True)

    return words_tuple

# word , count 
print("word , count")
for word in word_count():
    print("{} , {}".format(word[0], word[1]))
```
