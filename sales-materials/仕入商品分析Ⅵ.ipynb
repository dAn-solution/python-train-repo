```python
uriage_list = []
uriagedf_col_nm = uriagedf_n.columns
for i in range(len(uriagedf_n)):
    if uriagedf_n.iloc[i]['品目'] in us_list:
        uriage_list.append(uriagedf_n.iloc[i])
uriagedf_n = pd.DataFrame(uriage_list, columns=uriagedf_col_nm)
```

```python
# 品目CDをキーに大分類、小分類、品名、規格をセット
meisyou_dict = {}
for i in range(len(uriagedf_n)):
    if uriagedf_n.iloc[i]["品目"] in meisyou_dict:
        continue
    else:
        meisyou_dict[uriagedf_n.iloc[i]["品目"]] = [uriagedf_n.iloc[i]["大分類"], 
                                                    uriagedf_n.iloc[i]["小分類"], 
                                                    uriagedf_n.iloc[i]["品名"], 
                                                    uriagedf_n.iloc[i]["規格"]]
```

```python
# 品目毎に売上集計
uriage_group_dict = dict(uriagedf_n.groupby("品目").sum()["販売金額"])
num_dict = {}
for i in meisyou_dict.keys():
    num_dict[i] = [0,0,0,0,0,0,0,0,0]

# num_dictに売上金額をセット
for i, j in uriage_group_dict.items():
    num_dict[i][0] = j
```

```python
def sonota_num_set(num_dict):
    for i, j in num_dict.items():
        # 売上原価
        num_dict[i][4] = num_dict[i][1] + num_dict[i][3] - num_dict[i][2]
        # 粗利益
        num_dict[i][5] = num_dict[i][0] - num_dict[i][4]
        # 在庫回転率
        if ((num_dict[i][1]+num_dict[i][2])) == 0:
            num_dict[i][6] = 0
        else:
            num_dict[i][6] = num_dict[i][0] / ((num_dict[i][1]+num_dict[i][2])/2)
        # 在庫回転期間
        if num_dict[i][0] == 0:
            num_dict[i][7] = 0
        else:
            num_dict[i][7] = num_dict[i][2] / (num_dict[i][0] / 12)
        # 交差比率
        if ((num_dict[i][1]+num_dict[i][2])) == 0:
            num_dict[i][8] = 0
        else:
            num_dict[i][8] = num_dict[i][5] / ((num_dict[i][1]+num_dict[i][2])/2)
    
    return num_dict
```

```python
def siire_kingaku_set(num_dict, siiredf_n):
    """仕入金額をnum_dictにセット
    """
    for i in range(len(siiredf_n)):
        if siiredf_n.iloc[i]["品目"] in num_dict:
            num_dict[siiredf_n.iloc[i]["品目"]][3] += siiredf_n.iloc[i]["仕入金額"]
    
    return num_dict
```

```python
def zaiko_kingaku_set(num_dict, kisyudf_n, kimatudf_n):
    """期首在庫と期末在庫の金額をnum_dictにセット
    """
    for i in range(len(kisyudf_n)):
        if kisyudf_n.iloc[i]["品目"] in uriage_group_dict:
            num_dict[kisyudf_n.iloc[i]["品目"]][1] += kisyudf_n.iloc[i]["在庫金額"]
        else:
            num_dict[kisyudf_n.iloc[i]["品目"]] = [0,0,0,0,0,0,0,0,0]
            num_dict[kisyudf_n.iloc[i]["品目"]][1] += kisyudf_n.iloc[i]["在庫金額"]

    for i in range(len(kimatudf_n)):
        if kimatudf_n.iloc[i]["品目"] in uriage_group_dict:
            num_dict[kimatudf_n.iloc[i]["品目"]][2] += kimatudf_n.iloc[i]["在庫金額"]
        else:
            num_dict[kimatudf_n.iloc[i]["品目"]] = [0,0,0,0,0,0,0,0,0]
            num_dict[kimatudf_n.iloc[i]["品目"]][2] += kimatudf_n.iloc[i]["在庫金額"]
    
    return num_dict
```

```python
# 期首、期末在庫金額をnum_dictにセット[1][2]
num_dict = zaiko_kingaku_set(num_dict, kisyudf_n, kimatudf_n)
# 仕入金額をnum_dictにセット[3]
num_dict = siire_kingaku_set(num_dict, siiredf_n)
# その他経営数字をセット[4]～[8]
num_dict = sonota_num_set(num_dict)
```

```python
# setした各種経営数字をDataFrameに
meisyou_df = pd.DataFrame(meisyou_dict)
num_df = pd.DataFrame(num_dict)
concat_df = pd.concat([meisyou_df, num_df])
concat_df = concat_df.T
concat_df = concat_df.set_axis(["大分類","小分類","品名","規格","売上","期首在庫","期末在庫","仕入","売上原価","粗利益","在庫回転率","在庫回転期間","交差比率"], axis=1)
```

```python
concat_df["売上"] = concat_df["売上"].astype(int)
concat_df["期首在庫"] = concat_df["期首在庫"].astype(int)
concat_df["期末在庫"] = concat_df["期末在庫"].astype(int)
concat_df["仕入"] = concat_df["仕入"].astype(int)
concat_df["売上原価"] = concat_df["売上原価"].astype(int)
concat_df["粗利益"] = concat_df["粗利益"].astype(int)
concat_df["在庫回転率"] = concat_df["在庫回転率"].astype(float)
concat_df["在庫回転期間"] = concat_df["在庫回転期間"].astype(float)
concat_df["交差比率"] = concat_df["交差比率"].astype(float)
concat_df = concat_df.dropna()
concat_df = concat_df[concat_df["大分類"] != "_"]
```

```python
bar_data = concat_df[concat_df["期末在庫"] > 0]
bar_data_range = bar_data.copy()
bar_data_range["品目"] = bar_data_range.index
bar_data_range["品目"] = bar_data_range["品目"].astype(int)
columns_name = ["大分類","小分類","品目","品名","規格","売上","期首在庫","期末在庫","仕入","売上原価","粗利益","在庫回転率","在庫回転期間","交差比率"]
bar_data_range = bar_data_range[columns_name]
bar_data_range["粗利率"] = (bar_data_range["粗利益"] / bar_data_range["売上"]) * 100
bar_data_range = bar_data_range[bar_data_range["粗利益"] > 0]
bar_data_range["在庫回転率"] = round(bar_data_range["在庫回転率"], 2)
bar_data_range["粗利率"] = round(bar_data_range["粗利率"], 2)
```

```python
# 交差比率グラフ関数
def kousa_graph(tmp_data,t,t1,idx,fig_num):
    ax = fig.add_subplot(fig_list[idx])
    plt.title('交差比率：　{} {}'.format(t, t1), fontsize=18)
    sns.scatterplot(x="在庫回転率", y="粗利率", size="粗利益", sizes=(100, 500), hue_norm=(0, 7), legend="brief", data=tmp_data)
    # ラベル付け
    label_point(tmp_data["在庫回転率"], tmp_data["粗利率"], tmp_data["品目"], ax)
```

```python
# ポイント名称関数
def label_point(x,y,id,ax):
    df_tmp_tmp = pd.concat({'x':x,'y':y,'id':id}, axis='columns')
    df_tmp = df_tmp_tmp.copy()
    # adjustTextに渡すテキストラベルのリスト
    texts = []
    for i, point in df_tmp.iterrows():
        # text_plot = ax.text(point['x'], point['y'], int(point['id']), ha='center', va='center',fontsize=10)
        text_plot = ax.text(point['x'], point['y'], int(point['id']), ha='center', va='center',fontsize=9)
        texts.append(text_plot)
    # arrowpropsでラベルからプロットした点を指す線などの形状や色を指定
    adjust_text(texts)
    # adjust_text(texts, arrowprops=dict(arrowstyle='->', color='red'))
```

```python
# 交差比率
fig = plt.figure(figsize=(16, 20))
fig_list = []
g, k = divmod(len(daibunrui_nm), 2)
a1 = str(g + k)
for i in range(len(daibunrui_nm)):
    fig_num = int(a1 + "2" + str(i+1))
    fig_list.append(fig_num)

for idx, dai in enumerate(daibunrui_nm):
    tmp_data_tmp = bar_data_range[bar_data_range["大分類"]==dai]
    tmp_data = tmp_data_tmp.copy()
    tmp_data["品目"] = tmp_data["品目"].astype(int)
    t = dai
    t1 = nendo_nm

    kousa_graph(tmp_data, t, t1, idx,fig_list)

plt.tight_layout()
plt.grid
plt.show()
```
