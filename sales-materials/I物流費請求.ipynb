```python
# ファイル読込
df = pd.read_excel("/content/drive/MyDrive/Purchased_products/excel/4月輸入.xlsx", sheet_name=0)
```

```python
# BEGOデータを絞り込み
cd_list = [90332, 90396, 90406, 90392, 90393, 90463]
df = df[df['代理店CD'].isin(cd_list)]
```

```python
# usecol = ['代理店名', '輸送会社', '料金', '関税', '立替手数料', '燃油サーチャージ', '通関料', '(貨物)取扱料', 
#           '保税回送料', '航空運賃', '海上運賃', 'CY/CFS/ CHG/C.C', 'ターミナルチャージ/外貨搬出作業料', '入出庫料',
#           '（トラック）配達料', 'ドレージ', 'ブローカー管理費', '保税上屋荷役料', 'BL作成料/書類作成料',
#           '（税関検査）取扱手数料', 'セキュリティチャージ', '保険料', 'その他', '消費税']
usecol = ['代理店名', '輸送会社', '料金', '関税', '立替手数料', '燃油サーチャージ', '通関料', '(貨物)取扱料', 
          '保税回送料', '航空運賃', '海上運賃', 'CY/CFS/ CHG/C.C', 'ターミナルチャージ/搬出作業料', '入出庫料',
          '（トラック）配達料', 'ドレージ', 'ブローカー管理費', '保税上屋荷役料', 'BL作成料/書類作成料',
          '（税関検査）取扱手数料', 'セキュリティチャージ', '保険料', 'その他', '消費税']
```

```python
# nanを0に　dictのkeyを列名
df = df.fillna(0)
num = len(df)
dict = {}
for col in usecol:
    if col == '':
        dict[col] = list('代理店名' * len(df))
    else:
        dict[col] = list(np.zeros(len(df), dtype=int))
```

```python
# dictに金額セット
for i in range(len(df)):
    for col in usecol:
        if col == '料金':
            dict[col][i] += int(df.iloc[i][col]) - int(df.iloc[i]['割引料金'])
        else:
            if col == '代理店名':
                dict[col][i] = df.iloc[i][col]
            elif col == '輸送会社':
                dict[col][i] = df.iloc[i][col]
            else:
                dict[col][i] = int(df.iloc[i][col])
```

```python
dict
```

```python
# DataFrameを用意
claim_df = pd.DataFrame()
for col in usecol:
    claim_df[col] = dict[col]
```

```python
# 列名を配達日に
col_nm = []
for i in range(len(df)):
    date = df.iloc[i]['配達日']
    new_date = str(date.year)+'.'+str(date.month)+'.'+str(date.day)
    col_nm.append(new_date)
```

```python
col_nm
```

```python
claim_df = claim_df.T
claim_df.columns = col_nm
```
